Ans 1:

Data Collection: Gathering relevant data for modeling.
Data Cleaning: Preprocessing data to handle missing values, outliers, and inconsistencies.
Feature Engineering: Selecting, transforming, or creating new features from the data.
Data Splitting: Dividing data into training, validation, and test sets.
Model Selection: Choosing an appropriate machine learning algorithm.
Model Training: Training the selected model on the training data.
Model Evaluation: Assessing the model's performance using metrics like accuracy, precision, recall, etc.
    
Ans 2:

Structured Data: Tabular data with a well-defined schema, e.g., CSV files. Example: Customer purchase history in a spreadsheet.
Unstructured Data: Data without a specific form or organization, e.g., text, images. Example: Social media posts or images.
Semi-Structured Data: Data that is partially structured, e.g., JSON or XML files. Example: Website logs in JSON format.
        
Ans 3:

a. Numeric vs. Categorical Attributes: Numeric attributes represent numerical values and can be measured or ordered, whereas categorical attributes represent categories or labels and cannot be measured or ordered.

b. Feature Selection vs. Dimensionality Reduction: Feature selection involves selecting the most relevant features for the model, whereas dimensionality reduction involves reducing the number of features by transforming them into a lower-dimensional space while preserving as much information as possible.

Ans 4:

a. The Histogram:

A graphical representation of the distribution of a dataset.
Displays data frequency in intervals (bins) along a continuous range.
Helps visualize central tendency, dispersion, and skewness of data.
Useful for identifying patterns, outliers, and understanding data distribution.

b. Use a Scatter Plot:

A graphical tool to visualize the relationship between two numerical variables.
Each point represents an observation, with one variable on the x-axis and another on the y-axis.
Helps identify patterns like linear or non-linear relationships, clusters, or outliers.
Useful for correlation analysis, model validation, and exploratory data analysis.

c. PCA (Principal Component Analysis):

A dimensionality reduction technique that transforms original variables into a new set of variables (principal components) while preserving most of the data variance.
Helps in reducing data complexity, noise, and redundancy.
Useful for visualizing high-dimensional data, feature extraction, and improving model performance by reducing overfitting.
Requires data standardization before applying PCA to ensure variables are on the same scale.

Ans 5:

Investigating data is necessary to understand its characteristics, identify patterns, anomalies, or insights that may impact modeling outcomes.
There may be differences in exploring qualitative and quantitative data, as qualitative data may require subjective interpretation or different visualization techniques compared to quantitative data.

Ans 6:

Histogram shapes can vary, including bell-shaped (normal distribution), skewed (positive or negative), uniform, or multimodal distributions.
'Bins' in a histogram are intervals that divide the data range into discrete segments to count the frequency of data points within each segment.

Ans 7:

Data outliers can be identified using statistical methods like Z-score, IQR, or visual techniques like box plots or scatter plots.
Depending on the context and impact, outliers can be treated by removing, transforming, or imputing them, or using robust algorithms that are less sensitive to outliers.

Ans 8:

Central Inclination Measures: Measures like mean, median, and mode represent the central or typical value of a dataset.
Mean vs. Median: Mean can be influenced by outliers, leading to a significant variation from the median, which is less affected by extreme values, especially in skewed distributions.

Ans 9:

Scatter plots visualize the relationship between two variables, showing patterns like linear, non-linear, or no relationship.
Outliers can be identified as data points that deviate significantly from the overall pattern or trend in the scatter plot.

Ans 10:

Cross-tabs (contingency tables) display the frequency or distribution of data for two categorical variables, helping to identify relationships or dependencies between them, often visualized using heatmaps or bar charts.
