shutil.copy() vs. shutil.copytree():

shutil.copy(): This function is used to copy a single file from the source to the destination. It copies the file's content and its permissions.
shutil.copytree(): This function is used to recursively copy an entire directory tree from the source to the destination. It copies the directory and all its contents (sub-directories and files).

Difference between delete functions in send2trash and shutil modules:

send2trash: The send2trash module moves files and directories to the system's trash or recycle bin instead of permanently deleting them. This means you can recover them if needed.
shutil: The delete functions in the shutil module (shutil.rmtree() for directories and os.remove() for files) permanently delete files and directories, and they cannot be easily recovered.

Equivalent ZipFile method to File objects’ open() method:

The ZipFile method equivalent to the File objects' open() method is ZipFile.open(). It opens a file inside the ZIP archive and returns a file-like object that can be read or written.

```python
#Ans:5
import os
import shutil

def copy_files_with_extension(src_folder, dest_folder, extension):
    # Create destination folder if it doesn't exist
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)

    # Walk through the source folder tree
    for foldername, subfolders, filenames in os.walk(src_folder):
        for filename in filenames:
            if filename.endswith('.' + extension):
                file_path = os.path.join(foldername, filename)
                dest_path = os.path.join(dest_folder, filename)
                
                # Check if the source and destination paths are the same
                if file_path != dest_path:
                    shutil.copy(file_path, dest_folder)
                    print(f"Copied {file_path} to {dest_folder}")
                else:
                    print(f"Skipping {file_path} as it is the same as the destination")


```

```python
source_folder = 'C:\\Users\\abhis\\Downloads'
destination_folder = 'C:\\Users\\abhis\\Downloads\\Section 2\\Section 2'
file_extension = 'png'  # Change this to desired file extension
copy_files_with_extension(source_folder, destination_folder, file_extension)
```
