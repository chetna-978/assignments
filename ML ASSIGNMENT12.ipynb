Ans 1:
Prior probability is the initial probability assigned to a hypothesis before considering any new evidence.
Example: Before rolling a die, the prior probability of getting a '6' is 
1/6

Ans 2:
Posterior probability is the updated probability of a hypothesis after considering new evidence or data.
Example: After rolling a die and getting a '6', the posterior probability of having rolled a '6' is updated based on this new evidence.

Ans 3:
Likelihood probability represents the probability of observing the given data under a specific hypothesis.
Example: If a coin is believed to be fair, the likelihood of observing 3 heads in 5 tosses would be calculated based on the probability of getting heads with a fair coin.

Ans 4:
Naïve Bayes classifier is a probabilistic classifier based on Bayes' theorem with an assumption of independence between features. It is named 'Naïve' because it assumes that the presence (or absence) of a particular feature of a class is unrelated to the presence (or absence) of any other feature, given the class variable.

Ans 5:
Optimal Bayes classifier is a classifier that makes decisions based on the posterior probabilities and selects the class with the highest posterior probability.

Ans 6:
Two features of Bayesian learning methods are:

Incorporation of Prior Knowledge: Bayesian methods allow the incorporation of prior beliefs or information into the learning process.
Quantification of Uncertainty: Bayesian methods provide a framework for quantifying and updating uncertainties based on observed data.

Ans 7:
Consistent learners are learning algorithms that converge to the true hypothesis as the amount of training data increases.

Ans 8:
Two strengths of Bayes classifier are:

Probabilistic Framework: It provides a probabilistic framework that quantifies uncertainty and allows for a principled approach to decision-making.
Effective with Small Data: Bayes classifiers can perform well even with limited training data, especially when strong prior knowledge is available.

Ans 9:
Two weaknesses of Bayes classifier are:

Assumption of Feature Independence: The "naïve" assumption of feature independence may not hold in real-world scenarios, leading to suboptimal performance.
Sensitivity to Incorrect Assumptions: If the prior probabilities or the assumptions about the data distribution are incorrect, the classifier's performance may be adversely affected.

Ans 10:
a. Text classification: Naïve Bayes classifiers are commonly used for text classification tasks such as document categorization, spam detection, and sentiment analysis.
b. Spam filtering: In spam filtering, Naïve Bayes classifiers can be trained on labeled data to distinguish between spam and non-spam emails based on the presence or absence of specific words or features.
c. Market sentiment analysis: Naïve Bayes classifiers can analyze textual data from sources like social media to determine market sentiment, which can be used for predicting stock market trends or assessing public opinion.
