1. What exactly is []?
   - [] represents an empty list in Python.

2. How would you assign the value 'hello' as the third value in a list stored in the variable spam? (Assume [2, 4, 6, 8, 10] are in spam.)
   - To assign the value 'hello' as the third value in a list stored in the variable spam, you can use the assignment statement like this:
     ```
     spam[2] = 'hello'
     ```

3. Let's pretend the spam includes the list ['a', 'b', 'c', 'd'] for the next three queries.
   a. What is the value of spam[int(int('3' * 2) / 11)]?
      - The value would be 'd'. Here's the breakdown:
        `int('3' * 2)` evaluates to 33 (an integer). `33 / 11` evaluates to 3.0 (a float). Indexing with spam[3] retrieves the fourth element of the list, which is 'd'.
   
   b. What is the value of spam[-1]?
      - The value of spam[-1] is 'd'. Negative indices in lists count from the end, so -1 refers to the last element.

   c. What is the value of spam[:2]?
      - The value of spam[:2] is ['a', 'b']. It retrieves the elements from the beginning of the list up to (but not including) the element at index 2.

4. Let's pretend bacon has the list [3.14, 'cat', 11, 'cat', True] for the next three questions.
   a. What is the value of bacon.index('cat')?
      - The value of bacon.index('cat') is 1. It returns the index of the first occurrence of 'cat' in the list.
   
   b. How does bacon.append(99) change the look of the list value in bacon?
      - When you use `bacon.append(99)`, it adds the integer 99 to the end of the bacon list. Afterward, bacon will look like this: [3.14, 'cat', 11, 'cat', True, 99].

   c. How does bacon.remove('cat') change the look of the list in bacon?
      - When you use `bacon.remove('cat')`, it removes the first occurrence of 'cat' from the bacon list. Afterward, bacon will look like this: [3.14, 11, 'cat', True, 99].

5. What are the list concatenation and list replication operators?
   - List Concatenation Operator: `+` is used to combine two or more lists into a new list.
   - List Replication Operator: `*` is used to replicate a list a certain number of times to create a new list.

6. What is the difference between the list methods append() and insert()?
   - `append()`: Adds an element to the end of the list.
   - `insert()`: Inserts an element at a specified index in the list.

7. What are the two methods for removing items from a list?
   - `remove(item)`: Removes the first occurrence of the specified item from the list.
   - `pop(index)`: Removes and returns the item at the specified index from the list.

8. Describe how list values and string values are identical.
   - List values and string values are identical in that they can both contain a sequence of other values, and they support indexing and slicing. However, they are different in that list values are mutable (can be changed), whereas string values are immutable (cannot be changed once created).

9. What's the difference between tuples and lists?
   - Lists are mutable (can be changed), while tuples are immutable (cannot be changed after creation).
   - Lists are defined using square brackets [...], while tuples are defined using parentheses (...).
   - Lists support more methods for modification, whereas tuples have a smaller set of methods due to their immutability.

10. How do you type a tuple value that only contains the integer 42?
    - To type a tuple value containing only the integer 42, you can do it like this:
      ```
      my_tuple = (42,)
      ```

11. How do you get a list value's tuple form? How do you get a tuple value's list form?
    - To get a list value's tuple form, you can use the `tuple()` constructor:
      ```
      my_list = [1, 2, 3]
      my_tuple = tuple(my_list)
      ```
    - To get a tuple value's list form, you can use the `list()` constructor in a similar way.

12. Variables that "contain" list values are not necessarily lists themselves. Instead, what do they contain?
    - Variables that "contain" list values contain references or pointers to the memory location where the list data is stored.

13. How do you distinguish between copy.copy() and copy.deepcopy()?
    - `copy.copy()`: Used to create a shallow copy of an object, including its top-level elements. If the object contains nested objects (e.g., lists within lists), the inner objects are still references to the same memory locations.
    - `copy.deepcopy()`: Used to create a deep copy of an object, including all nested objects. It recursively creates copies of all nested objects, ensuring that the copied object is entirely independent of the original object.
