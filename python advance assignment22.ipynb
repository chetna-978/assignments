Q1. Benefits of the Built-in Array Package:

Benefits: The built-in array package in Python provides a more memory-efficient and performance-enhanced alternative to lists for certain operations. It is particularly useful when dealing with large datasets of homogeneous numeric data.

Q2. Limitations of the Array Package:

Limitations: The array package has limitations such as being less versatile compared to numpy arrays. It lacks many functionalities and operations available in numpy, which is a more comprehensive library for array manipulation.

Q3. Differences between Array and Numpy Packages:

Main Differences: Numpy is a more feature-rich and powerful library for numerical operations compared to the built-in array package. Numpy provides a broader range of functions, supports multidimensional arrays, and offers efficient operations for numerical computing.

Q4. Distinctions between empty, ones, and zeros Functions:

empty: Creates an array without initializing its values.
ones: Creates an array filled with ones.
zeros: Creates an array filled with zeros.

Q5. Role of the Callable Argument in fromfunction:

Role: The callable argument in numpy.fromfunction defines a function that takes coordinates as input and produces the value for each element in the new array. It allows for the creation of arrays based on a specified function.

Q6. Addition of Numpy Array with Single-Value Operand:

Outcome: When a numpy array is combined with a single-value operand through addition (e.g., A + n), the scalar value is added to each element of the array.

Q7. Array-to-Scalar Operations with Combined Operation-Assign Operators:

Outcome: Yes, array-to-scalar operations support combined operation-assign operators (e.g., += or *=). The operation is applied element-wise, modifying the original array in-place.

Q8. Numpy Array with Fixed-Length Strings:

String Type: Numpy arrays can contain fixed-length strings using the numpy.string_ data type.
Outcome: If a longer string is allocated to an array with fixed-length strings, the extra characters are truncated, and no error is raised.

Q9. Combining Two Numpy Arrays:

Conditions: Numpy arrays can be combined through addition or multiplication when they have compatible shapes. For addition, the arrays must have the same shape, and for multiplication, they need to conform to the rules of matrix multiplication or element-wise multiplication.

Q10. Using a Boolean Array to Mask Another Array:

Method: You can use boolean indexing. For example, masked_array = original_array[boolean_mask] will create a new array containing only the elements of the original array corresponding to True values in the boolean mask.

Q11. Getting Standard Deviation in Python:

Three ways:
Using Numpy: numpy.std(data)
Using Statistics Module: statistics.stdev(data)
Using Math Module (for Sample Standard Deviation): math.sqrt(sum((x - mean) ** 2 for x in data) / (len(data) - 1))

Q12. Dimensionality of Boolean Mask-Generated Array:

Dimensionality: The Boolean mask-generated array has the same dimensionality as the original array. It has True values at positions corresponding to True values in the mask and False elsewhere.
