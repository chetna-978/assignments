#2.
The difference between a string and a variable:

A string is a data type used to represent textual data, like words, sentences, or characters. For example, 'hello' is a string.
A variable is a symbolic name that represents a value or an object in a programming language. Variables can store different types of data, including strings, numbers, and more. For example, in the statement spam = 10, "spam" is a variable that stores the value 10.

#3.
Three different data types:

Integer: Represents whole numbers (e.g., 5, -3, 0).
Floating-point (float): Represents numbers with decimal points (e.g., 3.14, -0.5).
String: Represents text or characters enclosed in single (' ') or double (" ") quotes (e.g., 'hello', "world").

#4.
An expression is made up of operators and operands. It can be a combination of values and operators that, when evaluated, produce a result. Expressions perform operations like addition, subtraction, multiplication, and division. All expressions yield a value as their result.


#5.
The difference between an expression and a statement:

An expression is a combination of values and operators that can be evaluated to produce a result.
A statement is a complete line of code that performs an action. Assignment statements, like spam = 10, are examples of statements.
