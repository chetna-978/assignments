#1.
Functions are advantageous in programs for several reasons:

Modularity: Functions allow you to break your code into smaller, reusable pieces, making it easier to understand and maintain.
Reusability: Once a function is defined, it can be called multiple times from different parts of your program, promoting code reusability.
Abstraction: Functions hide the implementation details, allowing you to focus on high-level logic without needing to understand how a function works internally.
Readability: Functions can make your code more readable by giving meaningful names to blocks of code.
Debugging: Functions make it easier to identify and fix errors because you can isolate and test specific parts of your code.

#4.
The difference between a function and a function call is that a function is a block of code that defines a set of actions, while a function call is when you execute (run) the function to perform those actions.


#5.
In Python, there is one global scope for a program, which includes variables defined at the top level of the program. However, there can be multiple local scopes, such as within functions or methods, where variables are defined.


#6.
When a function call returns, the variables in its local scope are destroyed, and their values are typically no longer accessible.


#7.
A return value is the value that a function sends back to the caller. It is the result of the function's execution. Yes, it is possible to use a return value in an expression or assign it to a variable.


#8.
If a function does not have a return statement, it automatically returns None.


#9.
To make a function variable refer to a global variable, you can use the global keyword within the function. For example:
global_var = 10

def my_function():
    global global_var
    global_var += 1

my_function()
print(global_var)

#13.
To save a program from crashing when it encounters an error, you can use error handling techniques such as try-except blocks to catch and handle exceptions gracefully.


#14.
The purpose of the try clause is to enclose code that may raise exceptions. The purpose of the except clause is to specify what to do if an exception is raised within the try block. It allows you to handle exceptions or perform error-specific actions.
