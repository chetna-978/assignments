Quantitative vs. Qualitative Data:
Quantitative Data: Numerical data that can be measured or counted. Examples include age, weight, income, etc.
Qualitative Data: Categorical or descriptive data that represents characteristics or categories. Examples include gender, color, country, etc.
Distinction: Quantitative data can be expressed as numbers and subjected to mathematical operations, while qualitative data represents categories or labels and requires different analysis techniques.

Missing values can lead to biased or inaccurate models if not handled properly.

a. Data collected at regular intervals:

Refers to data points or measurements taken at consistent time or spatial intervals.
Often seen in time-series analysis, where data is recorded at fixed time intervals like daily, monthly, or yearly.
Useful for identifying patterns, trends, and seasonal variations in data.

Represents the range between the first quartile (Q1) and the third quartile (Q3) in a dataset.
Known as the interquartile range (IQR) and is a measure of statistical dispersion.
Useful for identifying variability and spread in the middle 50% of the data, especially when comparing different datasets or distributions.

c. Use a cross-tab:

A cross-tabulation (cross-tab) is a table that shows the frequency distribution of categorical variables.
Useful for analyzing relationships or dependencies between two or more categorical variables.
Helps in summarizing and visualizing data to understand patterns, associations, or trends between variables.

a. Comparison between Data with Nominal and Ordinal Values:

Nominal Data: Categorical data without any inherent order or ranking, like gender or color.
Ordinal Data: Categorical data with a meaningful order or ranking, but the intervals between categories may not be uniform, like rating scales (poor, average, good).
Difference: Nominal data can be categorized but not ranked, while ordinal data has a defined order or ranking but may lack uniform intervals between categories.

b. Histogram and Box Plot:

Histogram: A graphical representation of the distribution of numerical data, where data is grouped into bins and plotted as bars. It shows the frequency or count of data within each bin.
Box Plot: Also known as a whisker plot, it displays the distribution of data using quartiles. It shows the median, quartiles, and potential outliers in a dataset.
Difference: Histogram provides a visual representation of data distribution across bins, while a box plot summarizes the data distribution using key statistics and highlights outliers.

c. The Average and Median:

Average (Mean): Represents the central value calculated by summing all data values and dividing by the total number of values. It is sensitive to extreme values or outliers in the data.
Median: Represents the middle value when data is sorted in ascending or descending order. It is less affected by outliers and provides a robust measure of central tendency.
Difference: While both measures provide insights into central tendency, the average considers all data values equally, whereas the median focuses on the middle value, making it more robust to extreme values.
