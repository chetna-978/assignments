Ans 1: Machine learning is a subset of artificial intelligence (AI) that involves training algorithms to learn from data. Instead of being explicitly programmed to perform a task, these algorithms use statistical techniques to learn and make predictions or decisions.

Ans 2: Issues where machine learning shines include:

Medical diagnostics: Identifying diseases from medical images or predicting patient outcomes.
Natural language processing: Translating languages, generating text, or understanding sentiment in text.
Financial forecasting: Predicting stock prices or detecting fraudulent transactions.
Recommendation systems: Personalizing content recommendations in platforms like Netflix or Amazon.

Ans 3: A labeled training set consists of input data paired with the correct output or target. For instance, in an image recognition task, each image in the training set would have a label indicating what the image represents. Machine learning algorithms use this labeled data to learn the relationship between inputs and outputs.

Ans 4: Two important supervised tasks are:

Classification: Predicting a discrete label or category, such as whether an email is spam or not.
Regression: Predicting a continuous numerical value, such as predicting house prices based on features like size and location.

Ans 5: Examples of unsupervised tasks include:

Clustering: Grouping similar data points together, like segmenting customers into different groups based on purchasing behavior.
Dimensionality reduction: Reducing the number of variables in data while retaining its structure.
Anomaly detection: Identifying unusual patterns or outliers in data.
Association rule learning: Discovering interesting relations between variables in large datasets, like finding frequently purchased item combinations.

Ans 6: For making a robot walk through various unfamiliar terrains, a reinforcement learning model would be best suited. Reinforcement learning involves agents learning to make decisions by interacting with an environment to achieve a goal.

Ans 7: To divide customers into different groups, the k-means clustering algorithm would be suitable based on similarities in their purchasing behavior or preferences.

Ans 8: Spam detection is typically considered a supervised learning problem. Algorithms are trained on labeled data where each email is labeled as either spam or not spam, allowing them to learn patterns and make predictions on new, unseen emails.

Ans 9: An online learning system is a machine learning approach where the model continuously updates and learns from new data as it becomes available, rather than being trained on a fixed dataset.

Ans 10: Out-of-core learning refers to training machine learning models on data that is too large to fit into a computer's main memory. Unlike in-core learning, where data fits entirely in memory, out-of-core learning involves sequentially loading data batches for processing.

Ans 11: Similarity-based learning algorithms, like k-nearest neighbors (k-NN), make predictions using a similarity measure, classifying new data points based on the majority class of their nearest neighbors in the training set.

Ans 12:

Model Parameter: Parameters learned from data during training, such as weights in a neural network.
Hyperparameter: Parameters set prior to training that affect the learning process, such as the learning rate or number of hidden layers in a neural network.

Ans 13:

Criteria: Model-based algorithms look for patterns or relationships in data to build a predictive model.
Most popular method: Linear regression for continuous outcomes or logistic regression for classification tasks.
Making predictions: Once trained, model-based algorithms use the learned model to make predictions on new, unseen data.

Ans 14:

Overfitting: Where a model performs well on training data but fails to generalize to new data.
Underfitting: Where a model is too simple to capture underlying patterns in the data.
Data quality: Ensuring training data is accurate, relevant, and representative.
Computational resources: Managing large datasets and complex models that require significant computational power.

Ans 15:

Regularization: Techniques like L1 or L2 regularization to prevent overfitting.
Ensemble methods: Combining multiple models to improve generalization.
Data augmentation: Increasing training data size and diversity to help the model generalize better.

Ans 16: A test set is a separate portion of data reserved to evaluate a trained machine learning model's performance, assessing how well it might perform on unseen real-world data.

Ans 17: The purpose of a validation set is to tune hyperparameters and evaluate the model's performance during training, helping prevent overfitting by providing an independent dataset for model assessment.

Ans 18: A train-dev kit is a subset of training data used for model development and testing during the development phase. It aids in monitoring model performance on unseen data and identifying potential improvement areas.

Ans 19: Using the test set to tune hyperparameters can lead to overfitting, as the model's performance on the test set may become overly optimistic. This can result in poor generalization to new data if the model is overly tailored to the test set's specific characteristics.
