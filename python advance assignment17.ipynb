Q1. Greedy vs. Non-greedy Syntax:

Greedy: Matches as much as possible. Visual: .* (captures the entire string).
Non-greedy (Lazy): Matches as little as possible. Visual: .*? (captures the shortest possible substring).
Transforming: Add a ? after a quantifier to make it non-greedy.

Q2. Greedy vs. Non-greedy Impact:

Difference: Greedy grabs the longest match, non-greedy grabs the shortest.
Availability: If only a greedy match is available, it will be chosen.

Q3. Non-tagged Group in Simple Match:

Practical Difference: In a simple match without replacement, a non-tagged group may not make a practical difference.

Q4. Impact of Non-tagged Category:

Scenario: In complex patterns or replacement operations, using non-tagged categories may impact the program's outcomes by affecting grouping and capturing behavior.

Q5. Look-ahead Condition and Program Results:

Difference: Look-ahead does not consume characters.
Scenario: Useful when you want to check for a pattern without including it in the match, impacting program results.

Q6. Positive vs. Negative Look-ahead:

Positive Look-ahead ((?=...)): Ensures a pattern is followed.
Negative Look-ahead ((?!...)): Ensures a pattern is not followed.

Q7. Referring to Groups by Name:

Benefit: Enhances code readability and makes the regex more self-explanatory.

Q8. Identifying Repeated Items with Named Groups:

Example: Using named groups like (?P<animal>\w+), you can identify repeated items, e.g., finding repeated animals in "The cow jumped over the moon."
    
Q9. Scanner Interface vs. re.findall:

Scanner vs. re.findall: Scanner provides more control over parsing and tokenization, allowing additional actions during parsing.
    
Q10. Naming Convention for Scanner Object:

Naming Requirement: No, a Scanner object does not have to be named "scanner." You can choose any valid variable name.
