Ans 1:

Machine Learning: Machine learning is a subset of artificial intelligence (AI) that involves training algorithms to learn from data and make predictions or decisions without being explicitly programmed.
Applications in Business:
Predictive Analytics: Forecasting sales or customer behavior based on historical data to inform business strategies.
Fraud Detection: Identifying fraudulent transactions or activities in financial systems to mitigate risks.
Ethical Concerns: Potential biases in data, privacy issues, job displacement due to automation, and accountability for algorithmic decisions.

Ans 2:

i. Supervision of Experts: Learning directly from experts through guidance, feedback, and structured training programs.
ii. Indirect Assistance of Experts: Learning from expert resources such as books, online courses, or tutorials without direct interaction.
iii. Self-Education: Learning independently through self-study, experimentation, and practical experience.

Ans 3:

Supervised Learning: Regression, classification.
Unsupervised Learning: Clustering, dimensionality reduction.
Reinforcement Learning: Q-learning, policy gradients.

Ans 4:

Supervised Learning: Learns from labeled data.
Unsupervised Learning: Learns from unlabeled data.
Reinforcement Learning: Learns by interacting with an environment and receiving feedback.

Ans 5:

Well-Posed Learning Problem: A problem is well-posed if it has a unique solution, is stable with respect to small changes in input, and the solution depends continuously on the input data.
Characteristics: Clearly defined objectives, available data, appropriate algorithms, and measurable outcomes.

Ans 6:

Capability of Machine Learning: Machine learning is powerful but not a panacea. It excels in pattern recognition and prediction tasks but may struggle with complex reasoning, creativity, or tasks requiring human intuition.

Ans 7:

Methods and Technologies: Algorithms (e.g., neural networks, decision trees), frameworks (e.g., TensorFlow, PyTorch), platforms (e.g., Google Cloud AI, AWS SageMaker).
Defined in Detail:
Neural Networks: A network of interconnected nodes (neurons) that process data, commonly used in deep learning.
Decision Trees: A flowchart-like structure where each internal node represents a feature, each branch represents a decision, and each leaf node represents an outcome.

Ans 8:

Linear Regression: Predicts a continuous output based on input features, e.g., predicting house prices based on features like size and location.
Logistic Regression: Classifies data into discrete categories, e.g., classifying emails as spam or not spam based on content.

Ans 9:

Supervised Learning: Uses labeled data to learn patterns and make predictions, e.g., predicting stock prices.
Unsupervised Learning: Uses unlabeled data to identify patterns or structures, e.g., segmenting customers based on purchasing behavior.

Ans 10:

Machine Learning Process: Data collection, data preprocessing, model training, evaluation, deployment, monitoring, and iteration.

Ans 11:
a.Generalization vs. Abstraction: Generalization refers to applying learned knowledge to new situations, whereas abstraction involves representing complex data or concepts in a simplified form.

b.Guided Learning vs. Unsupervised Learning: Guided learning involves structured instruction or guidance, whereas unsupervised learning involves discovering patterns or structures in data without explicit guidance.

c.Regression vs. Classification: Regression predicts continuous numerical values, whereas classification categorizes data into discrete classes or labels.
