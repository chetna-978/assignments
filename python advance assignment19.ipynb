Q1. Relationship between a Class and its Instances:

The relationship between a class and its instances is a one-to-many partnership. A class serves as a blueprint or template, defining attributes and behaviors common to all instances. Instances are individual objects created from the class, each having its own unique state.

Q2. Data Held Only in an Instance:

Instance-specific data is held only in an instance. This includes attributes that vary between instances and are not shared among all objects of the class.

Q3. Knowledge Stored in a Class:

A class stores shared knowledge in the form of class variables and methods. Class variables are shared among all instances, while methods define behaviors that apply to all instances.

Q4. Method vs. Regular Function:

A method is a function associated with an object (instance or class). It is defined within a class and can access the object's attributes. A regular function is independent of any class or object and is not associated with a specific instance.

Q5. Inheritance in Python:

Yes, inheritance is supported in Python. The syntax for inheritance is achieved by placing the superclass (parent class) in parentheses after the subclass (child class). For example: class ChildClass(ParentClass):

Q6. Encapsulation in Python:

Python supports a limited form of encapsulation by using single or double underscores to make instance or class variables "private." However, it is based on convention, and variables are still accessible.

Q7. Distinguishing Class Variable and Instance Variable:

Class Variable: Shared among all instances of a class, defined outside methods.
Instance Variable: Specific to each instance, defined within methods using self.

Q8. Including "self" in Method Definitions:

"self" is always included as the first parameter in a class's method definitions. It represents the instance calling the method and allows access to its attributes and methods.

Q9. Difference between add and radd Methods:

_add_: Called for the operation a + b, where a is an instance of the class defining the method.
_radd_: Called for the operation b + a, allowing the right-hand object (b) to define the behavior.

Q10. Necessity of Reflection Method:

Reflection methods, like _getattr_ and _setattr_, are necessary when you need to customize attribute access or assignment dynamically. They provide control over attribute-based operations.

Q11. iadd Method:

_iadd_ is called for the in-place addition operation a += b. It allows the left-hand object (a) to define the behavior for in-place addition.

Q12. Inheritance of init Method:

Yes, the _init_ method is inherited by subclasses. If you need to customize its behavior within a subclass, you can override it by defining a new _init_ method in the subclass.**
