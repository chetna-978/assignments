The three different types of Boolean operators are:

AND (and)
OR (or)
NOT (not)

#3.Truth tables for Boolean operators:

AND (and):
    True and True = True
True and False = False
False and True = False
False and False = False

OR (or):
    
    True or True = True
True or False = True
False or True = True
False or False = False

NOT (not):
    
    not True = False
not False = True

#6.
The equal to (==) operator is used to compare two values for equality, while the assignment operator (=) is used to assign a value to a variable. You can differentiate them by context. For example:

x = 5 assigns the value 5 to the variable x.
if x == 5: checks if the value of x is equal to 5 in a conditional statement.

#10.
The break statement is used to exit a loop prematurely, while the continue statement is used to skip the rest of the current iteration and move to the next iteration of a loop.

#11.
In a for loop, range(10), range(0, 10), and range(0, 10, 1) are equivalent and will iterate from 0 to 9 (inclusive) with a step of 1.
