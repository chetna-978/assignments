Ans 1:

TensorFlow is an open-source machine learning library that allows for the creation and training of deep learning models. Its main features include scalability, flexibility, and a comprehensive ecosystem. Other popular Deep Learning libraries include PyTorch, Keras (which is now part of TensorFlow), and Theano.

Ans 2:

While TensorFlow offers functionalities similar to NumPy, it isn't a drop-in replacement. TensorFlow operates on tensors, and while it can perform many of the same operations as NumPy, the underlying structure and usage can differ. Additionally, TensorFlow is designed to work efficiently on GPUs.

Ans 3:

While both tf.range(10) and tf.constant(np.arange(10)) produce tensors with the same values, they may not be exactly identical due to potential precision differences or device placements. However, for most practical purposes, they are equivalent.

Ans 5:

Define a custom loss function using a function when the operation is simple and doesn't require stateful computations. Subclassing keras.losses.Loss is preferable when you need to maintain states across batches or require more complex operations.

Ans 6:

Define a custom metric using a function when the metric computation is straightforward. Subclassing keras.metrics.Metric is recommended when you need to maintain states or accumulate results across batches.

Ans 7:

Create a custom layer when you want to define a specific computation pattern that will be reused in multiple models. Create a custom model when you want to define the entire architecture or when you need to implement complex training loops or methods not available in standard Keras models.

Ans 8:

You might need to write your custom training loop for:
Implementing non-standard training methods like GAN training.
Incorporating specialized regularization techniques.
Implementing custom learning rate schedules or optimizers.
Handling advanced use-cases like distributed training.

Ans 9:

Custom Keras components must be convertible to TF Functions for graph mode execution. While they can contain arbitrary Python code during eager execution, this code needs to be encapsulated in a way that TensorFlow can trace and convert to a graph.

Ans 10:

Functions convertible to TF Functions should avoid:
Using Python constructs like lists or dictionaries.
Modifying global variables or states.
Accessing non-tensor objects or operations.
Using certain Python built-ins like print.

Ans 11:

You might need to create a dynamic Keras model when the architecture changes based on input data (e.g., conditional computation). To create a dynamic model, you would use the functional or subclassing API and introduce conditional branches based on input shapes or values. Making all models dynamic can reduce performance due to the inability to optimize computations in advance, so it's recommended only when necessary.
